```python
from datetime import timezone
import pytz

# Strip everything down to date and hour:minute
df['date_time_clean'] = df['started_at'].str.extract(r'(\d{4}-\d{2}-\d{2} \d{2}:\d{2})')

# Parse this simplified format
df['started_at_parsed'] = pd.to_datetime(df['date_time_clean'])

# Add UTC timezone info - handle ambiguous times by keeping them
df['started_at_utc'] = df['started_at_parsed'].dt.tz_localize('UTC')

# Convert to Chicago time with ambiguous time handling
chicago_tz = pytz.timezone('America/Chicago')
df['started_at_CT'] = df['started_at_utc'].dt.tz_convert(chicago_tz)

# Extract time features
df['hour'] = df['started_at_CT'].dt.hour
df['minute'] = df['started_at_CT'].dt.minute
df['day_of_week'] = df['started_at_CT'].dt.dayofweek
df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)

# For hour_only, instead of using floor which can cause DST issues,
# create a new timestamp from the date and hour components
df['date'] = df['started_at_CT'].dt.date
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.boxplot(df['trip_duration_minutes'])
plt.title('Box Plot of Trip Duration')
plt.ylabel('Trip Duration (minutes)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

percentiles = df['trip_duration_minutes'].describe()
print(percentiles)
```

```python
import numpy as np

# Filter out impossible values first (negative or zero)
filtered_df = df[df['trip_duration_minutes'] > 0]

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Create histogram with log scale on x-axis
# Use a reasonable number of bins to see the distribution clearly
bins = np.logspace(np.log10(filtered_df['trip_duration_minutes'].min()),
                  np.log10(filtered_df['trip_duration_minutes'].max()),
                  50)  # 50 bins on log scale

ax.hist(filtered_df['trip_duration_minutes'], bins=bins, alpha=0.7)
ax.set_xscale('log')  # Set x-axis to logarithmic scale

# Add grid for better readability
ax.grid(True, alpha=0.3, linestyle='--')

# Add vertical lines for key percentiles
percentiles = [25, 50, 75, 95]
for p in percentiles:
    value = np.percentile(filtered_df['trip_duration_minutes'], p)
    ax.axvline(x=value, color='r', linestyle='--', alpha=0.7,
               label=f'{p}th percentile: {value:.1f} min')

# Improve labels and title
ax.set_xlabel('Trip Duration (minutes) - Log Scale')
ax.set_ylabel('Frequency')
ax.set_title('Distribution of Bike Trip Durations (Log Scale)')

# Add legend
ax.legend()

# Show typical commuting range
commuter_min, commuter_max = 5, 30
ax.axvspan(commuter_min, commuter_max, alpha=0.2, color='green',
           label='Typical Commute Range')

# Show primary x-axis with actual values (not log values)
# This makes the plot easier to interpret
ticks = [1, 5, 10, 30, 60, 120, 240, 480, 1440]  # 1 min to 24 hours
ax.set_xticks(ticks)
ax.set_xticklabels([str(t) for t in ticks])

plt.tight_layout()
plt.show()

# Print summary statistics for key percentiles
print(filtered_df['trip_duration_minutes'].describe([0.05, 0.25, 0.5, 0.75, 0.90, 0.95, 0.99]))
```

```python
def categorize_trip(duration):
    if duration <= 20:
        return 'commuter'
    elif duration <= 45:
        return 'extended_commuter'
    else:
      return 'recreational'

df['trip_category'] = df['trip_duration_minutes'].apply(categorize_trip)

# Count the occurrences of each category
category_counts = df['trip_category'].value_counts()

# Display counts and percentages
print("Counts by trip category:")
for category, count in category_counts.items():
    percentage = (count / len(df)) * 100
    print(f"{category}: {count:,} ({percentage:.2f}%)")
```
